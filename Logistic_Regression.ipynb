```python
!kaggle datasets download -d fedesoriano/heart-failure-prediction
import pandas as pd
import numpy as np
df = pd.read_csv('/content/heart-failure-prediction.zip')
df
```

```python
heartDisease = df['HeartDisease']
# Droping data that is not in Numeric form
df = df.drop(['HeartDisease','ST_Slope','ExerciseAngina','RestingECG','Sex','ChestPainType'], axis=1)
df = df.to_numpy()
max = np.array([0,0,0,0,0,0])
# Feature Scaling
for i in range(df.shape[0]):
  if df[i][0] > max[0]:
    max[0] = df[i][0]
  if df[i][1] > max[1]:
    max[1] = df[i][1]
  if df[i][2] > max[2]:
    max[2] = df[i][2]
  if df[i][3] > max[3]:
    max[3] = df[i][3]
  if df[i][4] > max[4]:
    max[4] = df[i][4]
  if df[i][5] > max[5]:
    max[5] = df[i][5]

for i in range(df.shape[0]):
  for j in range(df.shape[1]):
    df[i,j] /= max[j]
df
```

```python
# Gradient Decent
def decent(heartDisease,df,w,b):
  m,n = df.shape
  dj_dw = np.zeros((n,))
  dj_db = 0
  for i in range(m):
    z = np.dot(df[i],w)+b
    output = 1/(1+np.exp(-z))
    error = output-heartDisease[i]
    for j in range(n):
      dj_dw[j] += error* df[i,j]
    dj_db += error
  return dj_dw/m,dj_db/m
```

```python
# Cost Function
def cost(heartDisease,X,w,b,max):
  m = X.shape[0]
  cost = 0
  for i in range(m):
    z = np.dot(X[i],w)+b
    output = 1/(1+np.exp(-z))
    cost += -heartDisease[i]*np.log(output) - (1-heartDisease[i])*np.log(1-output)
  return cost * (1/m)
```

```python
# Prediction
def predict(df,w,b):
  z = np.dot(df[0],w)+b
  output = 1/(1+np.exp(-z))
  return output
```

```python
w = np.array([1,1,1,1,1,1])
b = 1
a = 1
for x in range(100000):
  dj_dw,dj_db = decent(heartDisease,df,w,b)
  w = w-a*dj_dw
  b = b-a*dj_db
  print(x,"W:",w,"B:",b,"Cost:",cost(heartDisease,df,w,b,max),"Prerdict",predict(df,w,b),"Ans:",heartDisease[0] )
```
